```python
import numpy as np
import cv2
```

```python
# 展示图片
# import warnings
# warnings.simplefilter("error")
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator, FormatStrFormatter

def show_images(images,closize=2,set_locator=False,titles=[]):
    
    rows=int(len(images)/closize)+1
    fig=plt.figure(figsize=(34,14))
    
    xmajorLocator = MultipleLocator(100)  # 将x主刻度标签设置为50的倍数
    xminorLocator = MultipleLocator(50) #将x轴次刻度标签设置为5的倍数
    
    ymajorLocator = MultipleLocator(100)
    yminorLocator = MultipleLocator(50) #将x轴次刻度标签设置为5的倍数
    
    for find,image in enumerate(images):
        if not np.any(image>1):
            image=image*255
        ax=fig.add_subplot(rows,closize,find+1)
        ax.imshow(image)
        
        if len(titles)==0:
            plt.title('{}-({})'.format(find,image.shape)) #图片大小作为标题
        else:
            plt.title('{}-({})'.format(titles[find],image.shape)) #自定义标题
        
        if set_locator:
            #主刻度
            ax.xaxis.set_major_locator(xmajorLocator)
            ax.yaxis.set_major_locator(ymajorLocator)
            
            #次刻度
            ax.xaxis.set_minor_locator(xminorLocator)
            ax.yaxis.set_minor_locator(yminorLocator)
            
            ax.xaxis.grid(True, which='major',linestyle='-.') #x坐标轴的网格使用主刻度
            ax.yaxis.grid(True, which='major',linestyle='-.') #x坐标轴的网格使用主刻度
    plt.show()
```

```python
# 模拟图片1-空白图
demoimage=np.zeros((900,900,1),np.uint8)
show_images([demoimage],closize=1)
cv2.imwrite("demo-00.png",demoimage)
```

```python
# 模拟图片2-单个mask
demoimage=np.zeros((900,900,1),np.uint8)

cv2.rectangle(demoimage,(50,50),(200,250),255,-1)

show_images([demoimage],closize=1)
cv2.imwrite("demo-01.png",demoimage)
```

```python
# 模拟图片3-多个mask
demoimage=np.zeros((900,900,1),np.uint8)

cv2.rectangle(demoimage,(50,50),(400,400),255,-1)
cv2.rectangle(demoimage,(500,500),(800,800),255,-1)

show_images([demoimage],closize=1)
cv2.imwrite("demo-02.png",demoimage)
```

```python
# 模拟图片4-空洞mask
demoimage=np.zeros((900,900,1),np.uint8)

cv2.rectangle(demoimage,(50,50),(400,400),255,-1)
cv2.rectangle(demoimage,(100,100),(300,300),0,-1)

show_images([demoimage],closize=1)
cv2.imwrite("demo-03.png",demoimage)
```

```python
# 模拟图片5-内嵌mask
demoimage=np.zeros((900,900,1),np.uint8)

cv2.rectangle(demoimage,(50,50),(400,400),255,-1)
cv2.rectangle(demoimage,(100,100),(300,300),60,-1)

show_images([demoimage],closize=1)
cv2.imwrite("demo-04.png",demoimage)
```

```python
# 模拟图片5-2层内嵌mask
demoimage=np.zeros((900,900,1),np.uint8)

cv2.rectangle(demoimage,(50,50),(400,400),255,-1)
cv2.rectangle(demoimage,(100,100),(300,300),60,-1)
cv2.rectangle(demoimage,(100,100),(300,300),60,-1)

show_images([demoimage],closize=1)
cv2.imwrite("demo-05.png",demoimage)
```
